```python
import osmnx as ox
import networkx as nx
import numpy as np
from geopy.geocoders import Nominatim
import time
import folium
from folium.plugins import MarkerCluster, Search, AntPath
from folium.features import CustomIcon, DivIcon
import itertools
import requests
from networkx.algorithms import approximation as approx
import random
from shapely.geometry import LineString
import os

###################################
# 1. Setup & Geocoding
###################################
geolocator = Nominatim(user_agent="route_optimization_app")

def geocode_locations(delivery_points):
    """
    Geocode each location using Nominatim. The first item (index 0) is the warehouse.
    """
    geolocations = []
    for loc in delivery_points:
        try:
            geo = geolocator.geocode(loc, timeout=5)
            if geo:
                geolocations.append(geo)
                print(f"‚úÖ {loc} => Lat: {geo.latitude}, Lon: {geo.longitude}")
            else:
                print(f"‚ö†Ô∏è {loc} => No results from Nominatim.")
                geolocations.append(None)
        except Exception as e:
            print(f"‚ùå Error geocoding {loc}: {e}")
            geolocations.append(None)
        time.sleep(1.2)  # respect rate limits
    return geolocations

###################################
# 2. TomTom Real-Time Traffic
###################################
def fetch_tomtom_traffic(coordinates, tomtom_api_key):
    """
    Fetch real-time traffic from TomTom for each lat/lon.
    Returns dict {(lat, lon): speed_in_kmh}.
    """
    base_url = "https://api.tomtom.com/traffic/services/4/flowSegmentData/absolute/10/json"
    live_speeds = {}
    for (lat, lon) in coordinates:
        params = {
            "point": f"{lat},{lon}",
            "unit": "KMPH",
            "key": tomtom_api_key
        }
        try:
            resp = requests.get(base_url, params=params)
            resp.raise_for_status()
            data = resp.json()
            if "flowSegmentData" in data:
                seg = data["flowSegmentData"]
                speed = seg.get("currentSpeed", 1)
                live_speeds[(lat, lon)] = speed
                print(f"TomTom => ({lat},{lon}) | Speed: {speed} km/h")
            else:
                live_speeds[(lat, lon)] = 1
        except requests.exceptions.RequestException as e:
            print(f"Traffic Error @({lat},{lon}): {e}")
            live_speeds[(lat, lon)] = 1

    print(f"‚úÖ TomTom traffic data for {len(live_speeds)} coords.")
    return live_speeds

###################################
# 3. Define Delivery Points
###################################
# 0th => Warehouse
delivery_points = [
    "Dubai Mall, Dubai, UAE",   # Warehouse
    "Dubai Marina, Dubai, UAE",
    "Palm Jumeirah, Dubai, UAE",
    "Deira, Dubai, UAE",
    "Jumeirah Beach, Dubai, UAE",
    "Business Bay, Dubai, UAE",
    "Al Barsha, Dubai, UAE",
    "Dubai Airport, Dubai, UAE"
]

# Priority & Time Windows (with new locations)
time_priority_info = {
    "Dubai Mall, Dubai, UAE":   {"priority": None, "time_window": None},  # Warehouse
    "Dubai Marina, Dubai, UAE": {"priority": 1, "time_window": (9, 12)},
    "Palm Jumeirah, Dubai, UAE":{"priority": 2, "time_window": (10, 14)},
    "Deira, Dubai, UAE":        {"priority": 3, "time_window": (14, 18)},
    "Jumeirah Beach, Dubai, UAE":{"priority": 1, "time_window": (8, 11)},
    "Business Bay, Dubai, UAE": {"priority": 1, "time_window": (9, 15)},
    "Al Barsha, Dubai, UAE":    {"priority": 2, "time_window": (10, 16)},
    "Dubai Airport, Dubai, UAE":{"priority": 3, "time_window": (12, 20)}
}

###################################
# 4. OSMnx Graph
###################################
graph = ox.graph_from_place("Dubai, United Arab Emirates", network_type="drive")

###################################
# 5. Fuel & Emission Calculation
###################################
def calculate_fuel_and_emissions(distance_km, fuel_efficiency=12.0, fuel_price=3.0, emission_factor=0.2):
    """
    Returns (fuel_cost, co2_emission).
    """
    fuel_cost = (distance_km / fuel_efficiency) * fuel_price
    co2_emission = distance_km * emission_factor
    return fuel_cost, co2_emission

###################################
# 6. Update Graph with TomTom
###################################
def update_graph_with_tomtom(graph, live_speeds):
    for u, v, key, data in graph.edges(keys=True, data=True):
        if "geometry" in data and isinstance(data["geometry"], LineString):
            lat, lon = data["geometry"].coords[0]
        else:
            lat, lon = None, None
        if (lat, lon) in live_speeds:
            speed_kmh = live_speeds[(lat, lon)]
            try:
                if speed_kmh > 0:
                    travel_time_hr = data["length"] / (speed_kmh * 1000 / 3600)
                    data["travel_time"] = travel_time_hr * 1.2
                else:
                    data["travel_time"] = float("inf")
            except:
                pass
    print("‚úÖ Graph updated with TomTom-based travel_time.")
    return graph

###################################
# 7. TSP Solver (Single-Vehicle)
###################################
def find_optimal_route(graph, nodes):
    """
    Single-vehicle TSP with cost & emission factors, priority & time window penalties.
    """
    cost = {}
    # build cost matrix
    for i, j in itertools.permutations(range(len(nodes)), 2):
        try:
            base_time = nx.shortest_path_length(graph, nodes[i], nodes[j], weight="travel_time")

            dist_m = nx.shortest_path_length(graph, nodes[i], nodes[j], weight="length")
            dist_km = dist_m / 1000.0
            fuel_cost, co2_emission = calculate_fuel_and_emissions(dist_km)

            loc_j = delivery_points[j]
            pr_j  = time_priority_info[loc_j]["priority"]
            tw_j  = time_priority_info[loc_j]["time_window"]

            if pr_j is None:  # warehouse
                priority_penalty = 0
                time_penalty     = 0
            else:
                priority_penalty = (pr_j -1)*2.0
                arrival_time = 8.0 + base_time
                if tw_j:
                    start_t, end_t = tw_j
                    time_penalty = 15.0 if (arrival_time<start_t or arrival_time> end_t) else 0.0
                else:
                    time_penalty = 0.0

            # Weighted CO2 => co2_emission *0.5
            final_cost = base_time + fuel_cost + (co2_emission*0.5) + priority_penalty + time_penalty
            cost[(i,j)] = final_cost
        except nx.NetworkXNoPath:
            cost[(i,j)] = 99999

    # build complete graph
    K = nx.complete_graph(len(nodes))
    for (u,v) in K.edges():
        K[u][v]["weight"] = cost.get((u,v), float("inf"))
        K[v][u]["weight"] = cost.get((v,u), float("inf"))

    route_order = approx.greedy_tsp(K, weight="weight")
    if len(route_order)>1:
        route_order.append(route_order[0])

    print("üöÄ TSP route =>", route_order)
    return route_order

###################################
# Approach A: Splitting TSP Among Drivers
###################################
drivers = [
    {"id":1, "capacity":3, "color":"red"},
    {"id":2, "capacity":3, "color":"blue"},
    {"id":3, "capacity":3, "color":"orange"}
]

def assign_deliveries_to_drivers(tsp_order, drivers):
    """
    Splits the final TSP order among multiple drivers based on capacity.
    index=0 => warehouse; we keep it in route but chunk it accordingly
    """
    assigned_routes = []
    driver_idx = 0
    current_chunk = []
    cap = drivers[driver_idx]["capacity"]

    for drop in tsp_order:
        # if we exceed capacity => assign to the driver, then switch
        if len(current_chunk) >= cap:
            assigned_routes.append({"driver":drivers[driver_idx], "chunk": current_chunk})
            current_chunk = []
            driver_idx = (driver_idx+1) % len(drivers)
            cap = drivers[driver_idx]["capacity"]

        current_chunk.append(drop)

    # leftover
    if current_chunk:
        assigned_routes.append({"driver":drivers[driver_idx], "chunk": current_chunk})

    # debug
    for r in assigned_routes:
        print(f"Driver {r['driver']['id']} => route indices: {r['chunk']}")
    return assigned_routes

###################################
# MAIN
###################################
if __name__=="__main__":
    geolocations = geocode_locations(delivery_points)
    coords = [(g.latitude, g.longitude) for g in geolocations if g]

    if coords:
        # 1) TomTom speeds
        TOMTOM_API_KEY = "M4gbWbXRcVHKsmy42AesQzR2rlrUarfm"
        tomtom_speeds = fetch_tomtom_traffic(coords, TOMTOM_API_KEY)
        # 2) update graph
        graph = update_graph_with_tomtom(graph, tomtom_speeds)
        # 3) convert coords-> OSM nodes
        node_list = ox.distance.nearest_nodes(
            graph,
            [c[1] for c in coords],
            [c[0] for c in coords]
        )
        valid_nodes = [n for n in node_list if n in graph.nodes()]
        if len(valid_nodes) < len(node_list):
            node_list = valid_nodes

        # 4) solve single TSP
        tsp_order = find_optimal_route(graph, node_list)
        tsp_node_order = [node_list[i] for i in tsp_order]

        # Approach A => we chunk TSP route among drivers
        assigned_routes = assign_deliveries_to_drivers(tsp_order, drivers)

        # final path building
        def count_traffic_signals_in_path(path_nodes, G):
            signals=0
            for n in path_nodes:
                if G.nodes[n].get("highway")=="traffic_signals":
                    signals+=1
            return signals

        # create map
        m = folium.Map(tiles="CartoDB Positron")
        m.fit_bounds([(c[0], c[1]) for c in coords])

        # markers
        marker_fg = folium.FeatureGroup(name="Delivery Points").add_to(m)
        visited = set()
        priority_colors = {1:"red",2:"orange",3:"blue"}
        drop_number = 1

        for idx in tsp_order:
            if idx in visited:
                continue
            visited.add(idx)

            place = delivery_points[idx]
            latlng = coords[idx]

            if idx==0: # warehouse
                wh_offset= (latlng[0]+0.001, latlng[1])
                folium.Marker(
                    location=wh_offset,
                    tooltip=f"üè≠ Warehouse => {place}",
                    icon=DivIcon(
                        icon_size=(120,40),
                        icon_anchor=(0,0),
                        html='<div style="font-size:14px;color:black;font-weight:bold;">üè≠ Warehouse</div>'
                    )
                ).add_to(marker_fg)
            else:
                pr = time_priority_info[place]["priority"]
                colorp = priority_colors.get(pr,"gray")
                folium.Marker(
                    location=latlng,
                    tooltip=f"{drop_number} - {place}",
                    icon=folium.Icon(color=colorp, icon="thumbtack", prefix="fa")
                ).add_to(marker_fg)
                drop_number+=1

        # Build each driver route in sub-chunks
        for route_info in assigned_routes:
            driver = route_info["driver"]
            color_driver = driver["color"]
            chunk = route_info["chunk"]

            # build path
            path_osm = []
            for i in range(len(chunk)-1):
                start_idx = chunk[i]
                end_idx   = chunk[i+1]
                start_osm = tsp_node_order[start_idx]
                end_osm   = tsp_node_order[end_idx]
                if nx.has_path(graph, start_osm, end_osm):
                    subp = nx.shortest_path(graph, start_osm, end_osm, weight="length")
                    path_osm.extend(subp)

            if not path_osm:
                print(f"‚ö†Ô∏è Driver {driver['id']}: No route found for chunk {chunk}.")
                continue

            coords_path = [(graph.nodes[n]["y"], graph.nodes[n]["x"]) for n in path_osm]

            # compute total distance
            dist_m = 0
            for i in range(len(path_osm)-1):
                if graph.has_edge(path_osm[i], path_osm[i+1], 0):
                    dist_m += graph.edges[path_osm[i], path_osm[i+1], 0]["length"]
            dist_km = round(dist_m/1000,2)
            signals_count = count_traffic_signals_in_path(path_osm, graph)

            # approximate total time
            speed_city = 40.0 # km/h
            base_time_min = (dist_km / speed_city)*60
            time_signals  = signals_count*0.5
            total_time    = round(base_time_min + time_signals, 2)

            # total deliveries => len(chunk)-1 if chunk includes warehouse, or len(chunk) if not
            # let's do full chunk minus the warehouse if included
            # but if warehouse is in chunk => check the index:
            warehouse_in_chunk = 0 in chunk
            deliveries_in_leg  = len(chunk)-1 if warehouse_in_chunk else len(chunk)

            # total weight => placeholder
            total_weight = deliveries_in_leg * 5  # e.g. 5 kg each

            tooltip_txt = (
                f"Driver {driver['id']} Route\n"
                f"Total Distance: {dist_km} km\n"
                f"Est. Time: {total_time} min\n"
                f"Traffic Signals: {signals_count}\n"
                f"Deliveries: {deliveries_in_leg}\n"
                f"Weight: {total_weight} kg"
            )

            AntPath(
                locations=coords_path,
                dash_array=[10,20],
                delay=600,
                color=color_driver,
                weight=4,
                tooltip=tooltip_txt
            ).add_to(m)

        # Legend
        def add_legend(map_obj):
            legend_html = '''
            <div style="position: fixed; bottom: 50px; left: 50px; width: 250px; height: 180px; 
            background-color: white; z-index:9999; font-size:14px; border:2px solid grey; padding: 10px;">
            <b>Multi-Driver TSP (Approach A)</b><br>
            <i style="background: red; color: white; padding: 2px 5px;">D1</i> Driver 1<br>
            <i style="background: blue; color: white; padding: 2px 5px;">D2</i> Driver 2<br>
            <i style="background: orange; color: white; padding: 2px 5px;">D3</i> Driver 3<br>
            <hr style="margin:5px 0;">
            <i style="background: red; color: white; padding: 2px 5px;">1</i> Priority 1 (High)<br>
            <i style="background: orange; color: white; padding: 2px 5px;">2</i> Priority 2 (Med)<br>
            <i style="background: blue; color: white; padding: 2px 5px;">3</i> Priority 3 (Low)<br>
            </div>
            '''
            map_obj.get_root().html.add_child(folium.Element(legend_html))

        add_legend(m)

        # Disclaimer
        disclaimer_html = '''
            <div style="
                position: fixed; 
                bottom: 80px; 
                right: 10px; 
                width: 340px; 
                background-color: rgba(255, 255, 255, 0.95); 
                z-index:9999; 
                font-size:12px; 
                font-weight: bold; 
                padding: 12px;
                border-radius: 5px;
                border: 2px solid black;
                text-align: center;
                line-height: 1.2em;">
                üö¶ Approach A: Single TSP => Split Among Multiple Drivers<br/>
                Each driver has capacity=3 deliveries. The route is color-coded per driver.<br/>
                Cost & Emissions are considered in TSP cost function. 
            </div>
        '''
        m.get_root().html.add_child(folium.Element(disclaimer_html))

        # Hover effect
        hover_js = """
        <script>
            var elements = document.getElementsByClassName('leaflet-interactive');
            for (var i = 0; i < elements.length; i++) {
                elements[i].addEventListener('mouseover', function(event) {
                    event.target.setAttribute('stroke-width', '6');
                });
                elements[i].addEventListener('mouseout', function(event) {
                    event.target.setAttribute('stroke-width', '4');
                });
            }
        </script>
        """
        m.get_root().html.add_child(folium.Element(hover_js))

        m.save("ApproachA_multiDriver_split.html")
        print("‚úÖ Approach A: Single TSP + Splitting among drivers. Map => ApproachA_multiDriver_split.html")
    else:
        print("No valid coordinates found for the given addresses!")
```
